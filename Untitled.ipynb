```python
#import 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
#importing both data files and looking at them
train_data=pd.read_csv("train.csv")
train_data.describe()
train_data.sample(15)

```

```python
#for test data
test_data=pd.read_csv("test.csv")
test_data.describe
test_data.sample(10)
print(test_data.shape)
```

```python
#converting both these tables to matrices
train_data=pd.read_csv("train.csv").as_matrix()
print(train_data)
test_data=pd.read_csv("train.csv").as_matrix()
print(test_data)
```

```python
clf=DecisionTreeClassifier()#using sklearn's decision tree classifier
```

```python
#training data
xtrain=train_data[0:28000,1:]# all the columns starting from 1
train_label=train_data[0:28000,0]#only the first column as it contains the label 1,2,3,5,7,8..
```

```python
#clf fit takes list of features and labels
clf.fit(xtrain,train_label)
```

```python
#prediction data
xtest=test_data[0:28000,1:]
actual_label=train_data[0:28000,0]#used to predict

```

```python
d=xtest[15]#random 8th element,this is a row vector
d.shape=(28,28)#converting a row vector to 28,28 matrix
plt.imshow(255-d,cmap='gray')#plt is object of py plot
plt.show()
```

```python
#now learning model is ready for prediction lets play
print(clf.predict([xtest[9]]))#passing a single row vector,pass the feature to the label for the feature
plt.show()
```

```python
#measuring the acuracy
prediction=clf.predict(xtest)
count=0
for i in range (0,21000):
    count+=1 if prediction[i]==actual_label[i] else 0
print ("Accuracy=",((count/21000)*100))    
    
```
