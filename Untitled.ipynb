```python
#import 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
data=pd.read_csv("train.csv")
data.describe()
data.sample(15)

```

```python
data=pd.read_csv("train.csv").as_matrix()
print(data)
```

```python
clf=DecisionTreeClassifier()
```

```python
#training data
xtrain=data[0:21000,1:]
train_label=data[0:21000,0]
```

```python
#clf fit takes list of features and labels
clf.fit(xtrain,train_label)
```

```python
#prediction data
xtest=data[2100:,1:]
actual_label=data[21000:,0]#used to predict

```

```python
d=xtest[8]#random 8th element,this is a row vector
d.shape=(28,28)#converting a row matrix to 28,28 matrix
plt.imshow(255-d,cmap='gray')#pt is object of py plot
plt.show()
```

```python
#now learning model is ready for prediction lest play
print(clf.predict([xtest[9]]))#passing a single row vector,pass the feature to the label for the feature
plt.show()
```

```python
#measuring the acuracy
prediction=clf.predict(xtest)
count=0
for i in range (0,21000):
    count+=1 if prediction[i]==actual_label[i] else 0
print ("Accuracy=",((count/21000)*100))    
    
```
